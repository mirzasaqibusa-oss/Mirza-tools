```python
# ============================================================
# ✅ STEP 1 — Apni Email Daalo aur Access Check karo
# ============================================================

YOUR_EMAIL = "apni_email@gmail.com"  # 👈 Yahan sirf apni Gmail email likho

# ---- Aage kuch mat badlo ----
import urllib.request, json, sys

SHEET_ID = "13GxCB3yUkF3RrsBrVh2vqImSBgkjBc2SWA08Qzf21tg"

try:
    url = f"https://docs.google.com/spreadsheets/d/{SHEET_ID}/gviz/tq?tqx=out:json&sheet=Sheet1"
    with urllib.request.urlopen(url, timeout=10) as r:
        raw = r.read().decode()
        raw = raw[raw.index('{'): raw.rindex('}')+1]
        data = json.loads(raw)

    rows = data['table']['rows']

    # B1 = ON/OFF switch
    status_cell = rows[0]['c'][1]
    tool_on = status_cell and status_cell.get('v','').strip().upper() == 'ON'

    if not tool_on:
        print("🔴 " + "="*45)
        print("   Tool abhi band hai.")
        print("   MirZa SaQib ne service temporarily disable ki hui hai.")
        print("   Baad mein try karo.")
        print("="*47)
        sys.exit()

    # A column mein allowed emails (row 1 se)
    allowed = set()
    for row in rows[1:]:
        if row['c'][0] and row['c'][0].get('v'):
            allowed.add(row['c'][0]['v'].strip().lower())

    if YOUR_EMAIL.strip().lower() in allowed:
        print("✅ " + "="*45)
        print(f"   Access Granted! Welcome 🎉")
        print(f"   Email: {YOUR_EMAIL}")
        print("   Ab Step 2 cell chalao...")
        print("="*47)
    else:
        print("❌ " + "="*45)
        print("   Access Denied!")
        print(f"   '{YOUR_EMAIL}' authorized nahi hai.")
        print("   MirZa SaQib se rabta karo access lene ke liye.")
        print("="*47)
        sys.exit()

except SystemExit:
    raise
except Exception as e:
    print(f"⚠️ Connection error: {e}")
    print("Internet check karo ya thodi der baad try karo.")
    sys.exit()
```
